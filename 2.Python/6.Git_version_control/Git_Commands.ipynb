## Git commands

Here are the Git commands used in the image:
Work from github commndline tool like command prompt or bash terminal

1. `pwd`
   - Prints the current working directory. Better to work from a home directory eg users/bruhan for command prompt 
2. `ls`
   - Lists files and directories in the current directory.

3. `cd Documents/Python`
   - Changes the directory to `Documents/Python`.

4. `git init`
   - Initializes a new Git repository in the current directory.

5. `git add .`
   - Stages all changes (new files, modifications, deletions) in the current directory for the next commit.

6. `git commit -m "Initial commit"`
   - Commits the staged changes with the message "Initial commit".

7. `git status`
   - Shows the current status of the repository, indicating if there are any changes to commit.

8. `git log`
   - Displays the commit history for the repository.
