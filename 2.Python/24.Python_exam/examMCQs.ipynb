Qn5. What is the output for:
- S = [['him', 'sell'], [90, 28, 43]]
- S[0][1][1]


- '28'
- '8'
- 'i'
- 'e'

```python
# S is a list containing two elements:
# First element S[0] = ['him', 'sell']
# Second element S[1] = [90, 28, 43]

S = [['him', 'sell'], [90, 28, 43]]
print(S[0])        # ['him', 'sell']
print(S[0][1])     # 'sell'
print(S[0][1][1])  # 'e'
```
