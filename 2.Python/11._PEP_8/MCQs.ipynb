# 5.  What is the lambda expression required to filter out negative numbers from a list:


numbers = [-2, 5, -9, 8, -1, 10]
result = filter(, numbers)
print(list(result))

# Output objectives
lambda x: -x
lambda x: x < 0
lambda x: x >= 0
lambda x: x == 0

To filter out negative numbers from a list, we need a lambda expression that returns True for non-negative numbers (i.e., numbers greater than or equal to 0) and False for negative numbers. Let's analyze each option:

lambda x: -x
This simply negates each number, it doesn't filter anything.
lambda x: x < 0
This returns True for negative numbers, which is the opposite of what we want.
lambda x: x >= 0
This returns True for non-negative numbers (0 and positive numbers), which is exactly what we need.
lambda x: x == 0
This only returns True for zero, filtering out all other numbers.

The correct lambda expression to filter out negative numbers from a list is:


`lambda x: x >= 0`

This lambda function returns True for numbers that are greater than or equal to 0 (non-negative numbers), and False for negative numbers, effectively filtering them out.

````python
# Full implementation
numbers = [-2, 5, -9, 8, -1, 10]
result = filter(lambda x: x >= 0, numbers)
print(list(result))
````

`The error in the code is:`

`"The error is that circle.area and square.area are method references, not method calls."`

Explanation:
In Python, circle.area and square.area without parentheses refer to the method itself, not the result of calling the method. To call the method and get the area, you need to use parentheses: circle.area() and square.area().

````python
class Shape:
    def __init__(self, name):
        self.name = name

class Circle(Shape):
    def __init__(self, radius):
        super().__init__("Circle")
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius

class Square(Shape):
    def __init__(self, side):
        super().__init__("Square")
        self.side = side

    def area(self):
        return self.side * self.side

circle = Circle(3)
square = Square(4)
total_area = circle.area() + square.area()  # Corrected: method calls
print("Total Area:", total_area)
````
