# Regular expression involving  `\b` , findall() and sub() functions

In regular expressions, `\b` is an *anchor* that matches a *word boundary*. Word boundaries are positions in the text where a word starts or ends. Specifically, `\b` matches the position:

1. **Before the first character in a string**, if it's a word character (like at the start of the text).
2. **After the last character in a string**, if it's a word character (like at the end of the text).
3. **Between a word character and a non-word character**.

### What Counts as a Word Character?
Word characters include letters (`a-z`, `A-Z`), numbers (`0-9`), and the underscore (`_`). Non-word characters include spaces, punctuation marks, and symbols.

### Examples of `\b` in Action
1. **`\bword\b`**: Matches the word "word" exactly, only if it appears as a separate word. So, it would match:
   - "This is a word."
   - "Word is here."

   But it would not match:
   - "password" (because "word" is part of a larger word)

2. **`\bcat\b`**: Matches "cat" as a separate word but would not match "scatter" or "category."

3. **`cat\b`**: Matches "cat" only if it appears at the end of a word, so it would match:
   - "the cat" (at the end of the word)
   - "scat" (at the end of the word)
  
   But not:
   - "catapult" (because "cat" is not at the end of a word)

4. **`\b` at the start of a pattern**: Useful to match words at the beginning of a string or those surrounded by spaces, punctuation, or other non-word characters.

### Why Use `\b`?
`\b` helps ensure that you're matching whole words or word boundaries rather than just a string of characters within a larger word. It’s often used for `precise pattern matching, such as in search-and-replace operations` to avoid partial matches within longer words.

## Example 1: Matching a Whole Word
Let’s say we want to match the word "cat" as a standalone word and not as part of another word like "catalog" or "scattered."

```python
import re

text = "The cat sat on the catalog near a scattered cat."

# Regular expression using \b to match the exact word 'cat'
pattern = r"\bcat\b"

# Find all matches
matches = re.findall(pattern, text)

print("Matches:", matches)
```

## Explanation:
Here, \bcat\b matches "cat" only when it’s surrounded by a word boundary.

So, it finds "cat" in "The cat sat on..." and "scattered cat" but skips "catalog" and "scattered."

`However, the pattern is case insensitive and wont match any word in a case not specified`

```python
# Example 

import re

text = "The cat sat on the catalog near a scattered Cat."

# Regular expression using \b to match the exact word 'cat', 
# but doesnt match Cat becuase it has capital letter 'C' at the beginning of the word.
pattern = r"\bcat\b"

# Find all matches
matches = re.findall(pattern, text)

print("Matches:", matches)
```

## Example 2: Matching Words at the End
If we use cat\b (without the leading \b), it will match "cat" at the end of words, like in "scat."

```python
text = "The cat sat on the mat near a scatter and scat."

# Pattern to match words ending with 'cat'
pattern = r"cat\b"

matches = re.findall(pattern, text)

print("Matches:", matches)
```

`Explanation:`

cat\b finds "cat" at the end of both "cat" and "scat" but not "scatter" or "catalog."

This pattern is useful when you want to match words that end with a specific substring.

## Example 3: Matching Words Starting with a Specific Prefix
Let’s say we want to find all words in a sentence that start with "un":

```python
text = "He was uncertain about the unusual event in the underground."

# Pattern to match words that start with 'un'
pattern = r"\bun\w*"

matches = re.findall(pattern, text)

print("Matches:", matches)
```

`Explanation:`
- Here, \bun\w* matches any word that starts with "un."
- \b ensures the match starts at the beginning of a word.
- un matches the literal prefix.
- \w* allows any additional letters following "un" to complete the word.

## Example 4: Removing a Whole Word

Let’s say we want to remove all occurrences of the word "cat" from a sentence:


```python
# Pattern to remove the exact word 'cat'
text = "The cat sat on the mat near a scattered cat."
pattern = r"\bcat\b"

# Use re.sub() to replace all matches with an empty string

result = re.sub(pattern, "", text)

print("Result:", result)
```


`Explanation:`

- \bcat\b matches the word "cat" as a standalone word.

- re.sub() replaces all matches with an empty string, effectively removing the word "cat" from the sentence.

This is useful when you want to remove a specific word from a sentence.



## Example 5: Removing a Whole Word
Suppose we want to remove "test" only when it appears as a separate word, not when it’s part of "contest."

```python
text = "This is a test. And this is a contest."

# Using \b to remove 'test' as a whole word
pattern = r"\btest\b"
result = re.sub(pattern, "", text)

print("Result:", result)
```

### Explanation:
- re.sub() replaces instances of "test" that match \btest\b with an empty string, effectively removing it.
- Notice that "test" in "contest" is not removed because it’s part of a larger word.
### Summary of \b Usage
- \b helps match whole words or specific parts of words based on their position.
- It’s handy for searching or replacing words without accidentally matching parts of other words.

# The sub() function 

The sub() function in Python’s re (regular expression) module is used for substitution, which means replacing parts of a string that match a specified pattern with another specified string.

The basic syntax for sub() is:

```python
re.sub(pattern, replacement, string, count=0, flags=0)
```

# Here's a breakdown of its parameters:

- pattern: The regular expression pattern that defines what should be replaced.
- replacement: The string (or function) that you want to replace each match with.
- string: The original string in which replacements will be made.
- count: (optional) The maximum number of replacements to make. By default, count=0, meaning all occurrences are replaced.
- flags: (optional) Flags to modify the matching behavior, such as re.IGNORECASE for case-insensitive matching.

## Examples of Using re.sub()
- Example 1: Basic String Replacement
- Let's replace all instances of "cat" with "dog":

```python
import re

text = "The cat sat on the mat with another cat."

# Replace 'cat' with 'dog'
result = re.sub(r"cat", "dog", text)

print(result)
```

`Explanation`

Here, re.sub(r"cat", "dog", text) searches for "cat" in the text and replaces each instance with "dog."

`Example 2:` Using \b to Replace Whole Words Only

Let’s say we only want to replace "cat" when it appears as a standalone word, not as part of other words like "catalog."

```python
text = "The cat sat on the mat near a catalog."

# Replace 'cat' as a standalone word with 'dog'
result = re.sub(r"\bcat\b", "dog", text)

print(result)
```

`Explanation`

- \bcat\b ensures "cat" is replaced only when it’s a complete word, so "catalog" remains unchanged.

`Example 3:` Limiting the Number of Replacements with count

Suppose we only want to replace the first instance of "cat":

```python
text = "The cat sat on the mat with another cat."

# Replace only the first occurrence of 'cat' with 'dog'
result = re.sub(r"cat", "dog", text, count=1)

print(result)
```

`Explanation`

- By setting count=1, only the first "cat" is replaced with "dog."

`Example 4: Using a Function in replacement`

You can also pass a function as the replacement to dynamically determine what each replacement should be based on the matched text.

```python
text = "The price is $5. The price was $10 before."

# Function to double any matched number
def double_price(match):
    return str(int(match.group()) * 2)

# Replace prices with double their value
result = re.sub(r"\d+", double_price, text)

print(result)
```

`Explanation`

- The double_price function takes each matched number, converts it to an integer, doubles it, and returns it as a string.
- re.sub(r"\d+", double_price, text) applies this function to every numeric match (\d+ matches numbers).

`Summary`
- sub() is used to replace matching patterns in a string.
- You can replace based on simple patterns, whole words using \b, or even conditionally using functions.
