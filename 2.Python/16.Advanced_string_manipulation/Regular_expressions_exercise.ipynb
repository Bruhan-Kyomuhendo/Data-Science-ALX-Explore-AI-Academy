### Exercise 1

**Scenario**: Your first task is to analyse the hashtags used in these posts, as they can give insights into popular environmental campaigns.

**Exercise**: Write a Python function called `extract_hashtags` to extract all unique hashtags from the social media posts.

```python
# insert code here
import re
def extract_hashtags(text):
    # Use the re.findall() function to find all matches of the pattern
    hashtags = re.findall(r'#\w+', text)
    
    # Convert the list of hashtags to a set to remove duplicates
    unique_hashtags = set(hashtags)
    
    return unique_hashtags

# Test with the provided text
print(extract_hashtags(social_media_posts))
```

```python
# insert code here
import re
def extract_hashtags(text):
    # compile the regex pattern
    pattern = re.compile(r'#\w+')
    
    # Use the pattern to find all matches of the pattern
    hashtags = pattern.findall(text)
    
    # Convert the list of hashtags to a set to remove duplicates
    unique_hashtags = set(hashtags)
    
    return unique_hashtags

# Test with the provided text
print(extract_hashtags(social_media_posts))
```

### Exercise 4

**Scenario**: For geographical analysis, you need to extract mentioned locations such as "Amazon Rainforest" and "Madagascan rainforests".

**Exercise**: Write a function to extract proper names that refer to locations.

```python
import re

def extract_locations(text):
    # Compile the regular expression pattern for reuse
    pattern = re.compile(r'\b(?:[A-Z][a-z]*\s*){1,3}(?:Rainforest|rainforests|Forest|Amazon|Madagascan)\b')
    
    # Use the compiled pattern to find all location matches in the text
    locations = pattern.findall(text)
    return locations

# Call the function and display results
locations = extract_locations(social_media_posts)
print(locations)
```

### Exercise 4
1. **Regular expression pattern**: The regex `\b[A-Z][a-z]+(?:\s[A-Z][a-z]+)*(?:\s[rR]ainforests?)?\b` is designed to capture phrases that typically represent location names. It starts by matching any word that begins with a capital letter, which is common for proper nouns such as place names. The pattern allows for additional capitalised words to form multi-word names. The optional part `(?:\s[rR]ainforests?)?` is included to specifically capture terms like "rainforest" or "rainforests", which are relevant in the environmental context of our dataset.

2. **Filtering out false positives**: Chances are that the solution will include false positives like the standalone word "Rainforest". This would happen because the regex is general enough to capture any capitalised word, and "Rainforest" appeared in the text as part of a hashtag, which could lead to its unintended inclusion. To rectify this, we've added specific conditions in the filtering process:
   * not `loc.endswith(':')` excludes matches that end with a colon, as these are more likely to be hashtags or other non-location elements.
   * `'rainforest' in loc.lower()` ensures the phrase includes "rainforest" or "Rainforest", keeping the focus on relevant environmental terms.
   * `loc != 'Rainforest'` explicitly excludes the standalone word "Rainforest" to avoid this specific false positive.

```python
def extract_locations(text):
    # Regular expression to match potential location names
    pattern = r'\b[A-Z][a-z]+(?:\s[A-Z][a-z]+)*(?:\s[rR]ainforests?)?\b'

    # Find all matches
    potential_locations = re.findall(pattern, text)

    # Filtering based on context and avoiding false positives
    locations = [loc for loc in potential_locations if not loc.endswith(':') and 'rainforest' in loc.lower() and loc != 'Rainforest']

    return locations

# Test with the provided text
print(extract_locations(social_media_posts))
```

```python
# What each part means
\b(?:[A-Z][a-z]*\s*){1,3}(?:Rainforest|rainforests|Forest|Amazon|Madagascan)\b
```

Here's a breakdown of each part:

\b:

This represents a word boundary. It ensures that we only match whole words, preventing partial matches within other words.
- (?:[A-Z][a-z]*\s*){1,3}:

- (?: ... ) is a non-capturing group. This grouping allows us to apply repetition {1,3} to multiple elements without capturing the matched text separately.
- [A-Z] matches a single uppercase letter (A-Z), ensuring that the matched word starts with a capital letter, typical for proper nouns.
- [a-z]* matches zero or more lowercase letters following the uppercase letter, allowing it to match a full capitalized word.
- \s* matches zero or more whitespace characters, allowing spaces between words.
- {1,3} specifies that this non-capturing group should repeat 1 to 3 times. This means we’re looking for one to three capitalized words, which is typical for names like "Amazon Rainforest" or "Madagascan rainforests".

- (?:Rainforest|rainforests|Forest|Amazon|Madagascan):

Another non-capturing group that matches specific keywords associated with geographical locations.
- The `|` symbol inside the group acts as an OR operator, so it will match any of the following words:
- Rainforest
- rainforests (allows lowercase as well)
- Forest
- Amazon
- Madagascan
- These words are common in location names, so this part of the pattern helps refine our search to likely geographical terms.
- \b (at the end):

Another word boundary to mark the end of the matched location. It ensures that any other words following these locations are separated.

`Summary`
This pattern looks for one to three capitalized words followed by specific keywords (Rainforest, Forest, etc.) to capture geographical locations like "Amazon Rainforest" or "Madagascan rainforests" as complete phrases. 

-The use of \b at the beginning and end ensures that only whole words are matched, making it more precise.

- The (?: ... ) in regular expressions is used to create a non-capturing group.

- Explanation of (?: ... ):
- Normally, parentheses `( ... )` in a regex create a capturing group, meaning they capture the matched text inside them so you can retrieve it later.
- By adding `?:` at the beginning, you make it a `non-capturing group`. 
- `This means the group still applies any repetition or alternation inside it, but it won’t capture the matched text for later reference.
- Why Use Non-Capturing Groups?
`Non-capturing groups are useful when you want to:`

- Group parts of a pattern to apply quantifiers or alternation (|) without needing to store the matched text.
Reduce memory usage by not storing unnecessary matches, which can make your regex more efficient, especially with complex patterns.

`Example`

```python
import re

text = "Amazon Rainforest"

# Capturing group
capturing = re.search(r"(Amazon) Rainforest", text)
print("Capturing group:", capturing.groups())  # Output: ('Amazon',)

# Non-capturing group
non_capturing = re.search(r"(?:Amazon) Rainforest", text)
print("Non-capturing group:", non_capturing.groups())  # Output: ()
```

`In this example:`

- The capturing group (Amazon) captures and stores "Amazon" in the match result.
- The non-capturing group (?:Amazon) matches "Amazon" but doesn’t store it in the result.
- Using (?: ... ) keeps the pattern flexible without saving the matched content, which is useful for complex patterns where only specific portions need to be captured.
