1. Assignment Operators
Assignment operators are used to assign values to variables.

```python
# Assignment operators
a = 10   # Simple assignment
b = 5

a += b   # Equivalent to a = a + b
print ("Addition:", a )
a -= 3   # Equivalent to a = a - 3
print ("Subtraction:", a )
a *= 2   # Equivalent to a = a * 2
print ("Multiplication:", a )
a /= 4   # Equivalent to a = a / 4
print ("Division:", a)
print("Result of assignment operations:", a)
```

```python
''' or the variables can be added within the print function; values assigned to the variables will be used in 
the calculation'''
print ("Addition:", a + b )
```

- Modulus Assignment (%=)
The **modulus assignment operator (%=)** calculates the *remainder of a division* and assigns that remainder back to the variable. It is a shorthand way of writing x = x % y. This is useful when you want to keep the remainder of a division operation without explicitly using the modulus operator (%) each time.

```python
# Example
a = 10
a %= 3  # Equivalent to a = a % 3
print(a)  # Output will be 1, as 10 % 3 gives a remainder of 1
```

```python
print ("Hello, world")
```

In this case, a is first set to 10. Then, a %= 3 calculates the remainder when a is divided by 3 and assigns that remainder back to a. After the operation, a becomes 1.

- Exponentiation Assignment (**=)
The exponentiation assignment operator (**=) raises a variable to the power of another value and then assigns the result back to the variable. It is shorthand for writing x = x ** y. This is useful when you want to raise a variable to a certain power without writing the full expression

```python
# Example
b = 4
b **= 2  # Equivalent to b = b ** 2
print(b)  # Output will be 16, as 4 squared is 16
```

In this case, b is first set to 4. Then, b **= 2 raises b to the power of 2 (i.e., squares b) and assigns the result back to b. After the operation, b becomes 16.

2. Comparison Operators
Comparison operators are used to compare values.

```python
# Comparison operators
x = 10
y = 5

print("x == y:", x == y)  # Equals
print("x != y:", x != y)  # Not equals
print("x > y:", x > y)    # Greater than
print("x < y:", x < y)    # Less than
print("x >= y:", x >= y)  # Greater than or equal to
print("x <= y:", x <= y)  # Less than or equal to
```

3. Logical Operators
Logical operators combine conditional statements.

```python
# Logical operators
a = True
b = False

print("a and b:", a and b)  # Logical AND
print("a or b:", a or b)    # Logical OR
print("not a:", not a)      # Logical NOT
print("not b:", not b)      # Logical NOT
```

4. Bitwise Operators
Bitwise operators perform operations on the binary representations of integers

```python
# Bitwise operators perform operations at binary level
''' They manuplate bits in binary representations of a number, they treat operands (decimal or 
base 10 values) as sequences of binary numbers or bits'''
x = 6   # 6 = 0110 in binary
y = 3   # 3 = 0011 in binary

print("x & y:", x & y)   # Bitwise AND, represented by the ampersand (&) symbol
print("x | y:", x | y)   # Bitwise OR, represented by the vertical bar (|) or pipe symbol
print("x ^ y:", x ^ y)   # Bitwise XOR, represented by the caret (^) or circumflex  symbol
print("~x:", ~x)         # Bitwise NOT, represented by the tilde (~) symbol
print("x << 1:", x << 1) # Left shift, represented by the double less-than (<<) symbol
print("x >> 1:", x >> 1) # Right shift, represented by the double greater-than (>>) symbol

```

## Bitwise (&) AND operator
Returns **1** if both bits are **1**; else it returns **0**

## Example

```python
a = 10 
b = 5
''' NB; a and b are decimal (base 10) values, so they have to be converted to bits in order to use 
                  the & bitwise operator '''

''' a in binary is 1010, 
    b in binary is 0101'''
print ("Bitwise AND, i.e a & b :", a & b )
```

### In case we want to see what the computer is doing, we use the bin() function

```python
a = 10 
b = 5
bin(a & b)
```

```python
'''what happens in the background
  1010
& 0101
_______
  0000  = 0 (in decimal, this is what the computer prints on the screen for you to see)  '''
```
