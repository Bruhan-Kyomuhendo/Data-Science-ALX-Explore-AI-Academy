```python
class Car2:
    def __init__(self, brand, model, color, year):
        self.__brand = brand
        self.__model = model
        self.__color = color
        self.__year = year
        self.speed = 0

    def get_brand(self):
        return self.__brand

    def set_brand(self, brand):
        self.__brand = brand

    def get_model(self):
        return self.__model

    def set_model(self, model):
        self.__model = model

    def get_color(self):
        return self.__color

    def set_color(self, color):
        self.__color = color

    def get_year(self):
        return self.__year

    def set_year(self, year):
        self.__year = year

    def accelerate(self, value):
        self.speed += value
        print(f"{self.__brand} {self.__model} is accelerating. Speed is now {self.speed} km/h.")
    def reverse(self):
        self.speed = -self.speed
        print(f"{self.__brand} {self.__model} is reversing. Speed is now {self.speed} km/h.")

    def brake(self):
        self.speed = 0
        print(f"{self.__brand} {self.__model} is braking. Car has stopped.")

    def autonomous_navigation(self):
        print(f"{self.__brand} {self.__model} is navigating autonomously.")

    def obstacle_detection(self):
        print(f"{self.__brand} {self.__model} is detecting and avoiding obstacles.")

    def adaptive_decision_making(self):
        print(f"{self.__brand} {self.__model} is making adaptive decisions based on traffic conditions.")

    def environment_mapping(self):
        print(f"{self.__brand} {self.__model} is mapping its surroundings for precise navigation.")

    def vehicle_to_vehicle_communication(self):
        print(f"{self.__brand} {self.__model} is communicating with nearby vehicles to share road data.")

    def traffic_signal_recognition(self):
        print(f"{self.__brand} {self.__model} is recognizing and responding to traffic signals.")

    def driver_monitoring(self):
        print(f"{self.__brand} {self.__model} is monitoring the driver for safety.")

# Example usage
if __name__ == "__main__":
    my_car = Car2("Google", "Self-Driving Prototype", "White", 2025)
    my_car.autonomous_navigation()
    my_car.obstacle_detection()
    my_car.adaptive_decision_making()
    my_car.environment_mapping()
    my_car.vehicle_to_vehicle_communication()
    my_car.traffic_signal_recognition()
    my_car.driver_monitoring()
    my_car.accelerate(60)
    my_car.reverse()
    my_car.brake()
```

## Inheritance and Polymorphism
A self-driving car is a type of car, so it shares many **attributes/properties** and **methods** with the **`Car`** class, along with its own unique features. Instead of copying the entire **`Car`** class into the self-driving car class and modifying it, we can extend its functionality using inheritance and polymorphism. This promotes code reuse and simplifies code maintenance.

**Self driving car addtional attributes and methods**

**addtional attributes**
- sensors
- actuators
-

**additional methods**
- autonomously drive
- automatic brake

**Inheritance**


```python
class SelfDrivingCar(Car2):
  def __init__(self, brand, model, color, year, sensors):
    super().__init__(brand, model, color, year)
    self.__sensors = sensors

  def get_sensors(self):
    return self.__sensors

  def set_sensors(self, sensors):
    self.__sensors = sensors

  #polymorphism

  def brake(self):
    super().brake() # overriding the brake method of the parent class
    print("Activating automatic braking system...")

```

### Complete code Encapsulation and Abstraction

```python
# Encapsulation: We define a Car class with properties and methods that belong to the car
class Car:
    def __init__(self, brand, model, colour, year):
        self.__brand = brand  # Private attribute
        self.__model = model  # Private attribute
        self.__colour = colour # Private attribute
        self.__year = year    # Private attribute
        self.__speed = 0      # Default speed is 0

    # Getter methods for encapsulated data (Abstraction)
    def get_brand(self):
        return self.__brand

    def get_model(self):
        return self.__model
    
    def get_colour(self):
        return self.__colour

    def get_year(self):
        return self.__year

    # Set speed to mimic acceleration and deceleration
    def accelerate(self, value):
        self.__speed += value
        print(f"{self.__brand} {self.__model} is accelerating. Speed is now {self.__speed} km/h.")

    def brake(self):
        self.__speed = 0
        print(f"{self.__brand} {self.__model} is braking. Car has stopped.")
```

### Inheritance

```python
# Inheritance: SelfDrivingCar is a subclass of Car that inherits its attributes and methods
class SelfDrivingCar(Car):
    def __init__(self, brand, model, colour, year, sensors):
        super().__init__(brand, model, colour, year)  # Initialize parent class attributes
        self.sensors = sensors  # New attribute specific to driverless cars

     # Polymorphism: overriding the brake method for a self-driving car with enhanced functionality
    def brake(self):
        super().brake()
        print("Activating automatic braking system...")

    # New method to simulate the car driving autonomously
    def drive_autonomously(self, origin, destination):
        print(f" A {self.get_colour()} {self.get_brand()} {self.get_model()} is driving from {origin} to {destination} autonomously using {self.sensors} sensors.")

    # Method to simulate detecting obstacles
    def detect_obstacle(self):
        print(f"{self.get_brand()} {self.get_model()} detects an obstacle. Applying brakes!")
        self.brake()


```
