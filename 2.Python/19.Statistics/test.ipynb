````python
# complete code to make the plot 
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
%matplotlib inline

# Set parameters
mu = 0
variance = 2
sigma = np.sqrt(variance)

# Create x values
x = np.linspace(mu - 3*variance, mu + 3*variance, 100)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x, norm.pdf(x, mu, sigma), 'b-', label='Normal Distribution')

# Add labels and title
plt.title(f'Normal Distribution (μ={mu}, σ²={variance})')
plt.xlabel('x')
plt.ylabel('Probability Density')

# Add grid and legend
plt.grid(True, alpha=0.3)
plt.legend()

# Show plot
plt.show()
````

````python
import numpy as np
from scipy import stats

student_scores = np.array([55, 62, 67, 58, 63, 59, 61, 68, 60, 64])
mu = 60
x_bar = np.mean(student_scores)
s = np.std(student_scores, ddof=1)
n = len(student_scores)
alpha = 0.05

# Calculate t-statistic
t_stat = (x_bar - mu) / (s / np.sqrt(n))

# Calculate p-value
p_value = stats.t.cdf(t_stat, df=n-1) if t_stat < 0 else (1 - stats.t.cdf(t_stat, df=n-1))
p_value *= 2  # Two-tailed test

print("T statistic:", t_stat)
print("P-value:", p_value)
````
