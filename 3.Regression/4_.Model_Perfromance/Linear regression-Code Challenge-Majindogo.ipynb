Linear regression [Code Challenge]

In this code challenge, we analyse the relationship between various features and the standard yield of a farm to predict future yields. We utilise scatter plots and correlation coefficients to explore linear relationships, fit linear regression models, and evaluate these models using R-squared, MAE, MSE, and RMSE metrics. Additionally, we conduct residual analysis to assess the model's assumptions and improve its predictive accuracy.
