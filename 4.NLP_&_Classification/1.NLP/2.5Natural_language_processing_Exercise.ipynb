## Import libraries and read in the data

````python
import nltk
from nltk import TreebankWordTokenizer, SnowballStemmer
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
import string
import urllib

#nltk.download('wordnet')
#nltk.download('stopwords')
#nltk.download('omw-1.4')
````

The data used in this notebook is text from the book "Alice's Adventures in Wonderland" by Lewis Carroll. 

````python
# read in the data
def print_some_url():
    with urllib.request.urlopen('https://raw.githubusercontent.com/Explore-AI/Public-Data/master/Data/classification_sprint//alice_in_wonderland.txt') as f:
        return f.read().decode('ISO-8859-1')

data = print_some_url()
print(data[:863])
````

## Data preprocessing
We will first start by providing you with the functions required to remove punctuation, create a bag-of-words, and define a stemmer, tokeniser, and lemmatizer. Once you apply the functions to preprocess the data, you will be asked to perform some calculations and analysis in the exercise questions below.


**Convert to lowercase and remove punctuation** 

````python
#Function to remove punctuation

def remove_punctuation(words):
    words = words.lower()
    return ''.join([x for x in words if x not in string.punctuation])
````

````python
#Apply the remove_punctuation function to the data
data = remove_punctuation(data)
````

**Create a bag-of-words and assign our stemmer and lemmatizer**

````python
# Define stemmer function
stemmer = SnowballStemmer('english')

# Tokenise data
tokeniser = TreebankWordTokenizer()
tokens = tokeniser.tokenize(data)

# Define lemmatizer
lemmatizer = WordNetLemmatizer()

# Bag-of-words
def bag_of_words_count(words, word_dict={}):
    """ this function takes in a list of words and returns a dictionary 
        with each word as a key, and the value represents the number of 
        times that word appeared"""
    for word in words:
        if word in word_dict.keys():
            word_dict[word] += 1
        else:
            word_dict[word] = 1
    return word_dict

# Remove stop words
tokens_less_stopwords = [word for word in tokens if word not in stopwords.words('english')]

# Create bag-of-words
bag_of_words = bag_of_words_count(tokens_less_stopwords,{})
````

````python
def find_stem_and_lemma(tokens, n):
    """
    Finds the stem and lemma of the nth word in the token list.
    
    Parameters:
        tokens (list): List of tokenized words.
        n (int): Index of the word (1-based).
    
    Returns:
        dict: A dictionary with keys 'original', 'stem', and 'lemma'.
    """
    if n <= 0 or n > len(tokens):
        raise ValueError("Index out of range. Ensure n is within the bounds of the tokens list.")
    
    nth_word = tokens[n - 1]  # Convert to 0-based index
    return {
        'original': nth_word,
        'stem': stemmer.stem(nth_word),
        'lemma': lemmatizer.lemmatize(nth_word)
    }

# Use the function to find the 120th word
result = find_stem_and_lemma(tokens, 120)
print(result)
````

> what is zero based index
A zero-based index means that the numbering of elements in a sequence (like a list, array, or string) starts at 0 rather than 1.

For example, in a zero-based index system:

The first element has an index of 0.
The second element has an index of 1.
The third element has an index of 2, and so on.
This is the default indexing system in most programming languages like Python, C, and Java.

Example:



````python
tokens_list = ['apple', 'banana', 'cherry']

# Accessing elements using zero-based indexing:
print(tokens_list[0])  # Output: 'apple' (first element)
print(tokens_list[1])  # Output: 'banana' (second element)
print(tokens_list[2])  # Output: 'cherry' (third element)
````

When working with lists, always remember that index 0 corresponds to the first element. If you refer to the nth word in a human-readable 1-based system, you’ll need to subtract 1 to get the corresponding index in a zero-based system.

For example:

The 120th word in a list (1-based) would correspond to tokens[119] in zero-based indexing.

## Solutions

### Exercise 1

````python
def find_roots(token_list, n):
    
    root_dict = {}
    word = token_list[n-1]
    root_dict['original'] = word
    root_dict["stem"] = stemmer.stem(word)
    root_dict["lemma"] = lemmatizer.lemmatize(word)
    
    return root_dict
````

````python
find_roots(tokens, 120) 
````

### Exercise 2

````python
def count_stopwords(token_list):
    STOPwords = [word for word in token_list if word in stopwords.words("english")]
    return len(STOPwords)
````

````python
def most_common_word(bag, k):
    switch = [(value, key) for key, value in bag.items()]
    switch = sorted(switch)
    return switch[-k][1]
````

### Exercise 5

````python
def word_frequency_count(bag, n):
    total = sum(1 for value in bag.values() if value == n)
    return total
````
