## Good coding practices

## TigerBeetle style of programming


1. Just write code
2. Consider the inputs and outputs of the system … before you write code
3. Consider the inputs and outputs of subsystems (modules, classes, other units) … before you write code
4. Understand the data flow through the system (diagram, or written) … before you write code
5. Understand the data flow through the system over time. (e.g. sequence diagrams, etc) … before you write code
6. Formalize the properties by which inputs and outputs are correct (assertions, contracts, etc) … before you write code
7. Understand (next level, enforce) the invariants of the state of the system internal to the system (e.g. not at the interface boundaries) … before you write code
8. Formally model the system before you write any code using the invariants and other things discovered above … before you write code
9. Every additional step is going to reduce design errors, and you can always go back and iterate, but you’ll also end up needing to refactor in some cases…
10.  “Do not abbreviate variable names”
11. Assert a lower and upper bound on loop iterations.
